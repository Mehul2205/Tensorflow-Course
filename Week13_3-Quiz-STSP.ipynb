# Quiz

Q1. What is an example of a Univariate time series?


Ans. Hour by Hour Temperature

Q2. What is an example of a Multivariate time series?


Ans. Hour by Hour Weather

Q3. What is imputed data?

Ans. A projection of unknown (usually past or missing) data

Q4. A sound wave is a good example of time series data

Ans. True

Q5. What is Seasonality?

Ans. A regular change in shape of the data

Q6. What is a Trend?

Ans. An overall direction for data regardless of direction

Q7. In the context of time series, what is noise?

Ans. Unpredictable changes in time series data

Q8. What is AutoCorrelation?

Ans. Data that follow a predicatble shape,  even if sacle is different

Q9. What is a non-stationary time series?

Ans. One that has a disruptive event breaking trend and seasonality
